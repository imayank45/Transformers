```python
# import libraries
try:
  from transformers import GPT2LMHeadModel, GPT2Tokenizer
  print("Libraries imported successfully")
except:
  print("Error importing libraries")
```

```python
# load pretrained GPT-2 model and tokenizer
try:
  tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
  model = GPT2LMHeadModel.from_pretrained("gpt2")
  print("Model and tokenizer loaded successfully")
except:
  print("Error loading model and tokenizer")
```

```python
def generate_poem(prompt):
    try:
        # Encode the prompt text
        input_ids = tokenizer.encode(prompt, return_tensors="pt")

        # Generate the poem
        output = model.generate(
            input_ids,
            max_length=200,  # Use a reasonable max_length
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            num_return_sequences=1
        )

        # Decode the generated poem
        poem = tokenizer.decode(output[0], skip_special_tokens=True)
        return poem
    except Exception as e:
        print(f"Error in generating poem: {e}")
        return None# Prompt for poem generation
poem_prompt = """
Write a poem about the Generative AI Project Lifecycle:
From defining a scope so vast,
To models chosen, unsurpassed,
Through prompts and feedback, they refine,
Aligning outputs, step by step design.
"""

# Generate the poem
try:
    poem = generate_poem(poem_prompt)
    if poem:
        print("Generated Poem:\n")
        print(poem)
    else:
        print("Poem generation failed.")
except Exception as e:
    print(f"Error generating poem: {e}")
```
