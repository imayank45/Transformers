```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
def generate_product_description(product_name, features):
    try:
        prompt = f"""
        You are a creative marketing assistant specializing in writing engaging and detailed product descriptions.
        Your task is to generate a catchy and attractive description for the following product:

        Product Name: {product_name}
        Features: {features}

        Make the description engaging, professional, and persuasive.
        """
        input_ids = tokenizer.encode(prompt, return_tensors='pt')
        output = model.generate(
            input_ids,
            max_length=300,
            temperature=0.7,
            top_p=0.9,
            do_sample=True,
            num_return_sequences=1
        )
        return tokenizer.decode(output[0], skip_special_tokens=True)
    except Exception as e:
        return f"Error generating description: {e}"
```

```python

# Example usage
product_name = "Smart Coffee Mug"
features = "Keeps your coffee hot for 6 hours, sleek design, rechargeable battery"
description = generate_product_description(product_name, features)
print("Generated Product Description:\n")
print(description)
```
