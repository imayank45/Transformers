```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
```

```python
# generate story
def generate_story(prompt):

  # encode prompt
  input_ids = tokenizer.encode(prompt, return_tensors='pt')

  # generate story output
  output = model.generate(
      input_ids,
      max_length=200,
      num_return_sequences=1,
      temperature=0.8,
      top_p=0.95,
      do_sample=True
  )

  # decode and return story output
  return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
# prompt for story generation
story_prompt = """
Once upon a time, in the world of al, a team of engineers embarked on an exciting journey to build

1. They started by defining the scope, asking, "What problem can this assistant solve?" 2. Then they carefully selected the model, evaluating GPT-2, GPT-3, and other options.

3. with hard work, they adapted and aligned the model through fine-tuning and prompt engineering.

4. Finally, they integrated the model into an application that changed the way people worked
"""
```

```python
story = generate_story(story_prompt)
print('Generated story: \n')
print(story)
```
