```python
from transformers import MarianMTModel, MarianTokenizer

def translate_text(input_text, source_lang, target_lang):

  try:
    model_name = f'Helsinki-NLP/opus-mt-{source_lang}-{target_lang}'
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    model = MarianMTModel.from_pretrained(model_name)

    # tokenize input
    tokenize_text = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)

    # generate translation
    translated = model.generate(**tokenize_text)

    # decode translation
    translated_text = tokenizer.batch_decode(translated, skip_special_tokens=True)

    return translated_text
  except Exception as e:
    print(f"An error occurred: {e}")
    return None
```

```python
if __name__ == "__main__":
  print("Welcome to English Translator!")

  # taking input from transformer
  input_text = input("Enter english text you want to translate: ")
  print("\nSelect the target langugage: ")
  print("1. Hindi (hi)")
  print("2. German (de)")

  target_choice = input("Enter your choice (1 or 2): ").strip()

  # set target language based on user input
  source_lang = "en"
  if target_choice == "1":
    target_lang = "hi"
  elif target_choice == "2":
    target_lang = "de"
  else:
    print("Invalid choice. Please select 1 or 2.")
    exit()

  # translated text
  translated_text = translate_text(input_text, source_lang, target_lang)

  if translated_text:
    print(f"\nTranslated text ({target_lang}): {translated_text[0]}")
```

```python
def solve_addition(a,b):

  print(f"Problem: What is the sum of {a} and {b}")
  print("\nChain of thogh reasoning: ")
  steps=[]
  carry = 0
  result = []

  a_digits = list(map(int, str(a)))[::-1]
  b_digits = list(map(int, str(b)))[::-1]

  for i in range(max(len(a_digits), len(b_digits))):

    a_digit = a_digits[i] if i < len(a_digits) else 0
    b_digit = b_digits[i] if i < len(b_digits) else 0
    sum_digit = a_digit + b_digit + carry
    carry = sum_digit // 10
    result.append(sum_digit % 10)
    steps.append(f"Add {a_digit} + {b_digit} + {carry}: Write {result[-1]}, carry {carry}.")

  if carry > 0:
    result.append(carry)
    steps.append(f"Final carry {carry} added to the result")

  # display results
  for step in steps:
    print(step)

  # combine the results into final rsults
  final_result = int(''.join(map(str, result[::-1])))
  print(f"\nSolution: {a} + {b} = {final_result}")
```
